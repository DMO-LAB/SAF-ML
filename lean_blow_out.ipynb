```python
import numpy as np
from common import utils

fuelphypro = np.load('./lean_blow_out/New_Phy_properties.npy')
fuelchempro = np.load('./lean_blow_out/New_CHEM_properties.npy')
LBO = np.load('./lean_blow_out/lbo_300_450_550.npy')

# Compute the mean value of the array
mean_value_phy = np.nanmean(fuelphypro,axis=0)
mean_value_chem = np.nanmean(fuelchempro,axis=0)
mean_value_lbo = np.nanmean(LBO,axis=0)


# Replace all NaN values with the mean value
LBO = arr = np.nan_to_num(LBO, nan=mean_value_lbo)
fuelphypro = np.nan_to_num(fuelphypro,nan=mean_value_phy)
fuelchempro = np.nan_to_num(fuelchempro,nan=mean_value_chem)

# Instantiate the utils class
myutils = utils()
```

```python
properties = ['T90 (K)','T50 (K)','T10 (K)','surface Tension (mN/m)','Density(kg/m3)','Viscosity (cSt)','Smoke Point (MM)','Molecular Weight (kg/kmol)']
colors=['green','blue','black','red','orange','darkblue','grey','pink','green','blue','black','red','orange','darkblue','grey','pink','indigo','violet']
for i in range(3):
    y = LBO[:,i]
    T = ['300K','450k','550k']
    print("NEW SETS OF PLOT FOR T = "+T[i])
    print(".......................... \n"*5)
    myutils.showAllPlot(properties=properties,fuelpro=fuelphypro,y=y,y_label='LBO equivalence ratio @ '+T[i],colors=colors)
```

```python
properties_chem = ['H/C ratio','% Aromatics','% iso-paraffins',' %n-paraffins',' %c-paraffins','DCN','Flash Point(K)', 'Heat of Combustion MJ/kg']
for i in range(3):
    y = LBO[:,i]
    T = ['300K','450k','550k']
    print("NEW SETS OF PLOT FOR T = "+T[i])
    print(".......................... \n"*5)
    myutils.showAllPlot(properties=properties_chem,fuelpro=fuelchempro,y=y,y_label='LBO equivalence ratio @ '+T[i],colors=colors)
```
