### Loading the SAF physical and chemical properties as well as the data for the high altitude relight

```python
#  Helper function
import matplotlib.pyplot as plt
import numpy as np
from common import utils
import seaborn as sns
#  Load the properties of the fuel
fuelphypro = np.load('./highaltitude_relight/fuels_PHY_properties.npy')
fuelchempro = np.load('./highaltitude_relight/fuels_CHEM_properties.npy')

# Loading the LBO
LBO = np.load('./highaltitude_relight/equi_ratio.npy')

# Temperature of the experiment
T = "239"

# List of physical properties considered
properties = ['Density @15 dg Celsius (kg/m3)','Surface Tension (mN/m)','Viscosity @20 dg Celsius (cSt)','Smoke Point (mm)','Flash Point(dg Celsius)','T10','T50']

# Color representing each fuel
colors=['green','blue','black','red','orange','darkblue','grey','pink']

# Load the helper object
myutils = utils()

```

```python
#  Show relationship between each physical property and LBO (During high altitude relight)
y = LBO

myutils.showAllPlot(properties=properties,fuelpro=fuelphypro,y=y,y_label='equi. ratio @25% ignition probability',colors=colors)
```

```python
propertieschem = ['%H content (mass%)','%Aromatic content (wt%)','%n-paraffins (wt%)','%i-paraffins (wt%)','%c-paraffins (wt%)','Molecular Weight (g/mole)','Derived Cetane Number','Heat Combustion (MJ/kg)']

myutils.showAllPlot(properties=propertieschem,fuelpro=fuelchempro,y=y,y_label='equi. ratio @25% ignition probability',colors=colors)
```

### Showing Physical properties correlation

```python
myutils.showCorrelation(fuelpro=fuelphypro,properties=properties)
```

### Showing Chemical properties correlation

```python
myutils.showCorrelation(fuelpro=fuelchempro,properties=propertieschem)
```

### Showing the feature importance plot when considering only the physical properties

```python
myutils.showFeatureImportance(y=y,fuelpro=fuelphypro,properties=properties,title='Feature importance ')
```

### Showing the feature importance plot when considering only the chemical properties

```python
myutils.showFeatureImportance(y=y,fuelpro=fuelchempro,properties=propertieschem,title='Feature importance ')
```

### Show combine chemical and physical properties correlation

```python
combinepro = properties + propertieschem
datacomb = np.concatenate((fuelphypro,fuelchempro),axis=1)
myutils.showCorrelation(fuelpro=datacomb,properties=combinepro)
```
